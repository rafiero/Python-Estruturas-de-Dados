#Busca em Profundidade

Regra 1: Visite um no adjacente nao visitado, o marque e o empilhe

Regra 2: Se nao puder seguir a regra 1, desempilhe um no

Regra 3: Se nao puder seguir a regra 1 ou a regra 2, terminou

Analogia: Labirinto utilizando uma corda

Precisa definir um ponto de inicio

##Pilha

```
import numpy as np
class Pilha:
  def __init__(self, capacidade):
    self.__capacidade = capacidade
    self.__topo = -1

    # Mudança do tipo do array
    self.__valores = np.empty(self.__capacidade, dtype=object)

  def __pilha_cheia(self):
    if self.__topo == self.__capacidade - 1:
      return True
    else:
      return False

  def __pilha_vazia(self):
    if self.__topo == -1:
      return True
    else:
      return False

  def empilhar(self, valor):
    if self.__pilha_cheia():
      print('A pilha está cheia')
    else:
      self.__topo += 1
      self.__valores[self.__topo] = valor

  def desempilhar(self):
    # Retorna o elemento desempilhado
    if self.__pilha_vazia():
      print('A pilha está vazia')
      return None
    else:
      temp = self.__valores[self.__topo]
      self.__topo -= 1
      return temp

  def topo(self):
    if self.__topo != -1:
      return self.__valores[self.__topo]
    else:
      return -1
```

#Implementacao

```
pilha = Pilha(5)
pilha.empilhar(grafo.arad)
pilha.empilhar(grafo.sibiu)
pilha.empilhar(grafo.timissoara)
```

```
class BuscaProfundidade:
  def __init__(self, inicio):
    #inicio vai ser um objto da classe vertice
    self.inicio = inicio
    self.inicio.visitado = True
    self.pilha = Pilha(20)
    self.pilha.empilhar(inicio)

  def buscar(self):
    topo = self.pilha.topo()
    print('Topo: {}'.format(topo.rotulo))
    #a cada loop que ele encontra um no adjacente nao visitado a busca entra em uma chamada recursiva
    for adjacente in topo.adjacentes:
      print('_____Topo eh {}. {} jah foi visitada? {}'.format(topo.rotulo, adjacente.vertice.rotulo, adjacente.vertice.visitado))
      if adjacente.vertice.visitado == False:
        adjacente.vertice.visitado = True
        self.pilha.empilhar(adjacente.vertice)
        print('_____Empilhou {}'.format(adjacente.vertice.rotulo))
        self.buscar()
    #a condicao de parada das chamadas recursivas e completar o loop e nao ter mais nenhum noh adjacente nao visitado
    print('Desempilhou: {}'.format(self.pilha.desempilhar().rotulo))
    print()

```

```
busca_profundidade = BuscaProfundidade(grafo.arad)
```

```
busca_profundidade.buscar()
```

#Busca em Largura

Regra 1: visita um no adjacente nao visitado, marque-o e coloque-o na fila

Regra 2: se nao puder fazer a regra 1, retire um no da fila e torne-o o no atual

Regra 3: se nao puder realizar a regra 1 ou 2, terminou


```
import numpy as np
class FilaCircular:
  
  def __init__(self, capacidade):
    self.capacidade = capacidade
    self.inicio = 0
    self.final = -1
    self.numero_elementos = 0
    
    # Mudança no tipo de dado
    self.valores = np.empty(self.capacidade, dtype=object)

  def __fila_vazia(self):
    return self.numero_elementos == 0

  def __fila_cheia(self):
    return self.numero_elementos == self.capacidade

  def enfileirar(self, valor):
    if self.__fila_cheia():
      print('A fila está cheia')
      return
    
    if self.final == self.capacidade - 1:
      self.final = -1
    self.final += 1
    self.valores[self.final] = valor
    self.numero_elementos += 1

  def desenfileirar(self):
    if self.__fila_vazia():
      print('A fila já está vazia')
      return

    temp = self.valores[self.inicio]
    self.inicio += 1
    if self.inicio == self.capacidade - 1:
      self.inicio = 0
    self.numero_elementos -= 1
    return temp

  def primeiro(self):
    if self.__fila_vazia():
      return -1
    return self.valores[self.inicio]
```

```
class BuscaLargura:
  def __init__(self, inicio):
    self.inicio = inicio
    self.inicio.visitado = True
    self.fila = FilaCircular(20)
    self.fila.enfileirar(inicio)

  def buscar(self):
    primeiro = self.fila.primeiro()
    print('--------------')
    print('Primeiro da fila: {}'.format(primeiro.rotulo))
    temp = self.fila.desenfileirar()
    print('Desenfileirou: {}'.format(temp.rotulo))
    for adjacente in primeiro.adjacentes:
      print('Primeiro era {}. {} jah foi visitado? {}'.format(temp.rotulo, adjacente.vertice.rotulo, adjacente.vertice.visitado))
      if adjacente.vertice.visitado == False:
        adjacente.vertice.visitado = True
        self.fila.enfileirar(adjacente.vertice)
        print('Enfileirou: {}'.format(adjacente.vertice.rotulo))
    if self.fila.numero_elementos > 0:
      self.buscar()
```

```
busca_largura = BuscaLargura(grafo.arad)
```

```
busca_largura.buscar()
```
