# Lista encadeada simples

```
class No:
  def __init__(self, valor):
    self.valor = valor
    self.proximo = None #nao existe NULL no python
  
  def mostra_no(self):
    print(self.valor)
```

```
class Pilha:
  def __init__(self):
    self.primeiro = None

  def __vazia(self):
    return self.primeiro == None

  def empilhar(self, valor):
    novo = No(valor)
    novo.proximo = self.primeiro
    self.primeiro = novo

  def desempilhar(self):
    if self.__vazia():
      print('A lista está vazia')
      return None
    
    temp = self.primeiro
    self.primeiro = self.primeiro.proximo
    return temp

  def mostrar(self):
    if self.primeiro == None:
      print('A lista está vazia')
      return None
    else:
      return self.primeiro.valor

  
```

```
pilha = Pilha()
```

```
pilha.empilhar(10)
```

```
pilha.empilhar(100)
```

```
pilha.empilhar(1000)
pilha.empilhar(10000)
```

```
pilha.mostrar()
```

```
pilha.desempilhar()
```

```
pilha.mostrar()
```
