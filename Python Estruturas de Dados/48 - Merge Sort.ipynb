#Merge Sort

### Baseado na tecnica dividir para conquistar
### Ordenacao por juncao


```
import random
```

```
def mergesort(lista, inicio=0, fim=None):
  if fim is None:
    fim = len(lista)
  if (fim - inicio > 1): #poderia usar len
    meio = (fim+inicio)//2
    mergesort(lista, inicio, meio)
    mergesort(lista, meio, fim)
    merge(lista, inicio, meio, fim)
  return lista
  

def merge(lista, inicio, meio, fim):
  left = lista[inicio:meio]
  right =  lista[meio:fim]
  i = j = 0
  for k in range(inicio, fim):
    if i >= len(left):
      lista[k] = right[j]
      j+=1
    elif j >= len(right):
      lista[k] = left[i]
      i+=1
    elif left[i] < right[j]:
      lista[k] = left[i]
      i+=1
    else:
      lista[k] = right[j]
      j+=1
    


```

```
lista = random.sample(range(1, 1000),20)
```

```
mergesort(lista)
```
