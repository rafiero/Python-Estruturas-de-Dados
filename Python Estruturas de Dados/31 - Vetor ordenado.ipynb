# Vetor ordenado

```
import numpy as np
```

```
class VetorOrdenado:
  
  def __init__(self, capacidade):
    self.capacidade = capacidade
    self.ultima_posicao = -1
    self.valores = np.empty(self.capacidade, dtype=int)

  # O(n)
  def imprime(self):
    if self.ultima_posicao == -1:
      print('O vetor está vazio')
    else:
      for i in range(self.ultima_posicao + 1):
        print(i, ' - ', self.valores[i])

  # O(n)
  def insere(self, valor):
    if self.ultima_posicao == self.capacidade - 1:
      print('Capacidade máxima atingida')
      return
    #encontra posicao para alocar valor
    posicao = 0
    for i in range(self.ultima_posicao + 1):
      if self.valores[i] > valor:
        posicao = i
        break
      if i == self.ultima_posicao:
        posicao = i + 1
    #rearranja valores maiores p direita, para liberar espaco da posicao
    x = self.ultima_posicao
    while x >= posicao:
      self.valores[x + 1] = self.valores[x]
      x -= 1
    #aloca no vetor
    self.valores[posicao] = valor
    self.ultima_posicao += 1

  # O(n)
  def pesquisa_linear(self, valor):
    for i in range(self.ultima_posicao + 1):
      if valor < self.valores[i]:
        return -1
      if valor == self.valores[i]:
        return i
    return -1 #fim do vetor

  # O(log n)
  def pesquisa_binaria(self, valor):
    esquerda = 0
    direita = self.ultima_posicao
    
    while True:
      meio = int((esquerda + direita)/2)
      #achou
      if self.valores[meio] == valor:
        return meio
      #nao encontrou o valor
      elif esquerda > direita:
        return -1
      else:
        #limite inferior
        if self.valores[meio] < valor:
          esquerda = meio + 1        
        #limite superior
        else:
          direita = meio -1  

  # O(n)
  def excluir(self, valor):
    posicao = self.pesquisa_linear(valor)
    if posicao == -1:
      return -1
    else:
      for i in range(posicao, self.ultima_posicao):
        self.valores[i] = self.valores[i + 1]
      
      self.ultima_posicao -= 1
```
