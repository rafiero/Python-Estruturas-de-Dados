# Expressoes Regulares

Padroes de busca em textos, como palavras que comecam com letras maisuculas, CEPs, emails, numero de telefone, etc

Existem metodos de se trabalhar com expressoes regulares


*   Search - Encontrar posicoes de padroes em uma string, se estes estiverem presentes
*   Match - Encontrar se o comeco da string eh igual a um determinado padrao
*   Findall - Encontrar todas as substrings em uma string que correspondam a um padrao




# Metacaracteres

*  . - indica qq caractere (exceto linha nova)
*  \w - indica qq caractere alfanumerico
*  \W - indica qq caractere nao-alfanumerico
*  \d - indica qq caractere que seja um digito (0-9)
*  \D - indica qq caractere que nao seja um digito (0-9)
*  \s - indica espaco em branco
*  ^ - indica que uma palavra deve comecar com determinada letra
*  $ - indica que uma palavra deve terminar com determinada letra
*  " \ " - usado antes de metacarecteres para especificar seu significado literal


## Quantificadores

permitem determinar como e quantas vezes os metacarecteres aparecem

*  [] - opcional entre os que estao dentro dos colchetes
*  () - captura de grupos de caracteres
*  * - de zero a mais vezes
*  ? - zero ou uma vez 
*  + - uma ou mais vezes
*  {m, n} - de m a n **vezes**



# Exemplo

expressao regular que detecta emails
