# Recursão

## Mostrar mensagens

```
for i in range(5):
  print(i, 'Recursão')
```

```
def recursao(i):
  if i > 5:
    return
  print(i, "resursao")
  return recursao(i+1)
```

```
recursao(1)
```

## Somar valores

```
def soma1(n):
  soma = 0
  for i in range(n + 1):
    soma += i
  return soma
```

```
soma1(5)
```

```
def soma2(n):
  if n == 0:
    return 0

  return n + soma2(n - 1)
```

```
soma2(5)
```

```
%timeit soma1(100)
```

```
%timeit soma2(100)
```
