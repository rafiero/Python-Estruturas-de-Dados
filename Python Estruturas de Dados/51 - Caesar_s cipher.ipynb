```
def new_character(character, key):
  alphabet = ['a','b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm',
            'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']  
  return alphabet[(alphabet.index(character) + key)%len(alphabet)]
 
def encrypt(msg, key):
  encrypted = ""
  for character in msg:
    encrypted = encrypted + new_character(character, key)
  return encrypted
```

```
key = int(input("Type an integer: "))
msg = input("Type a message to be encrypted: ")
```

```
encrypted = encrypt(msg, key)
print("Encrypt: ", encrypted)
```
