# Fila lista encadeada simples

```
class No:
  def __init__(self, valor):
    self.valor = valor
    self.proximo = None #nao existe NULL no python
  
  def mostra_no(self):
    print(self.valor)
```

```
class Fila:
  def __init__(self):
    self.primeiro = None
    self.ultimo = None

  def __vazia(self):
    return self.primeiro == None

  def enfileirar(self, valor):
    novo = No(valor)
    if self.__vazia():
      self.primeiro = novo
      self.ultimo = novo
    self.ultimo.proximo = novo
    self.ultimo = novo
    return True
    

  def desenfileirar(self):
    if self.__vazia():
      print('A lista está vazia')
      return None
    
    self.primeiro = self.primeiro.proximo
    return True

  def mostrar(self):
    if self.primeiro == None:
      print('A lista está vazia')
      return None
    else:
      return self.primeiro.valor

  
```

```
fila = Fila()
```

```
fila.enfileirar(10)
```

```
fila.enfileirar(100)
fila.enfileirar(1000)
fila.enfileirar(10000)
fila.enfileirar(100000)
fila.enfileirar(1000000)
```

```
fila.mostrar()
```

```
fila.desenfileirar()
fila.desenfileirar()
fila.desenfileirar()
```

```
fila.mostrar()
```
