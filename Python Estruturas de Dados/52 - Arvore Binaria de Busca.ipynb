```
class No:
  def __init__(self, valor):
    self.valor = valor
    self.esquerda = None
    self.direita = None
  
  def mostra_no(self):
    print(self.valor)
```

```
class arvoreBinariaBusca:
  def __init__(self):
    self.raiz = None
    self.ligacoes = []
  
  def inserir(self, valor):
    #1 - O local a ser inserido deve ser encontrado
    #2 - Segue-se o caminho da raiz ate o devido no, que sera pai do novo no
    #3 - Quando o pai for localizado, o novo no sera conectado como seu filho, a esquerda ou a direita
    #4 - BigO: O(logn) caso medio, pior caso O(n)
    novo = No(valor)
    #Se a arvore estiver vazia
    if self.raiz == None:
      self.raiz = novo
    else:
      atual = self.raiz
      while True:
        pai = atual
        #esquerda
        if valor < atual.valor:
          atual = atual.esquerda
          if atual == None:
            pai.esquerda = novo
            self.ligacoes.append(str(pai.valor)+'->'+str(novo.valor))
            return
        #direita
        else:
          atual = atual.direita
          if atual == None:
            pai.direita = novo
            self.ligacoes.append(str(pai.valor)+'->'+str(novo.valor))
            return

  #Visualizacao da arvore feita pela webgraphviz.com

  def pesquisa(self, valor):
    if self.raiz == None:
      print("Arvore Vazia")
      return
    else:
      atual = self.raiz
      while atual.valor != valor:
        if valor < atual.valor:
          atual = atual.esquerda
        else:
          atual = atual.direita
        if atual == None:
          return None
      return atual
  #raiz, esquerda, direita
  def preOrdem(self, no):
    if no != None:
      print(no.valor)
      self.preOrdem(no.esquerda)
      self.preOrdem(no.direita)

  #esquerda, raiz, direita
  def ordem(self, no):
    if no != None:
      self.ordem(no.esquerda)
      print(no.valor)
      self.ordem(no.direita)
  
  #esquerda, direita, raiz
  def posOrdem(self, no):
    if no != None:
      self.posOrdem(no.esquerda)     
      self.posOrdem(no.direita)
      print(no.valor)
    

  def excluir(self, valor):
    if self.raiz == None:
      return    
    #encontrar o no
    atual = self.raiz
    pai = self.raiz
    ehEsquerda = True

    while atual.valor != valor:
      pai = atual
      #esquerda
      if valor < atual.valor:
        ehEsquerda = True
        atual = atual.esquerda
      #direita
      else:
        ehEsquerda = False
        atual = atual.dereita
      #n existe elemento buscado
      if atual == None: 
        return False
      
    #O no eh Folha
    if atual.esquerda == None and atual.direita == None:
      if atual == self.raiz:
        self.raiz == None
      elif ehEsquerda == True:        
        pai.esquerda = None
      else:
        pai.direita = None
      #apaga elemento do vetor usado para visualiazacao
      self.ligacoes.remove(str(pai.valor)+'->'+str(atual.valor))
    
    #O no tem um filho
    #filho a esquerda
    elif atual.direita == None:
      if atual == self.raiz:
        self.raiz == atual.esquerda
      elif ehEsquerda == True:
        pai.esquerda = atual.esquerda
      else:
        pai.direita = atual.esquerda
    #filho a direita
    elif atual.esquerda == None:
      if atual == self.raiz:
        self.raiz == atual.direita
      elif ehEsquerda == True:
        pai.esquerda = atual.direita
      else:
        pai.direita = atual.direita

    #O no tem dois filhos
    else:
      sucessor = self.getSucessor(atual)
      if(atual == self.raiz):
        self.raiz = sucessor
      elif ehEsquerda == True:
        pai.esquerda = sucessor
      else:
        pai.direita = sucessor

      sucessor.esquerda = atual.esquerda

    return True
    

       
  def getSucessor(self, no):
    paiSucessor = no
    sucessor = no
    atual = no.direita
    
    while atual != None:
      paiSucessor = sucessor
      sucessor = atual
      atual = atual.esquerda
    if sucessor != no.direita:
      paiSucessor.esquerda = sucessor.direita
      sucessor.direita = no.direita
    return sucessor




  

    
```
