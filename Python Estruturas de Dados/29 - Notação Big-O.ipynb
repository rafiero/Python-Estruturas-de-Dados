# Notação Big-O

- Como comparar dois algoritmos?
- Comparação objetiva entre algoritmos, sem levar em concideracao hardware ou linguagem usada
- Considera diferenças entre poder de processamento, sistema operacional, linguagem de programação
- O quanto a "complexidade" do algoritmo aumenta de acordo com as entradas

## Função 1 - O(n)

```
# 11 passos
def soma1(n):
  soma = 0
  for i in range(n + 1):
    soma += i
  
  return soma
```

```
soma1(10)
```

```
%timeit soma1(10)
```

## Função 2

```
(10 * 11) / 2
```

```
# 3 passos
# O(3)
def soma2(n):
  return (n * (n + 1)) / 2
```

```
soma2(10)
```

```
%timeit soma2(10)
```

- https://wiki.python.org/moin/TimeComplexity

```
# O(1000) -> O(n)
def lista1():
  lista = []
  for i in range(1000):
    lista += [i] #nesse caso voce esta criando uma lista de um elemento com i
    #lista.append(i)
  return lista
```

```
print(lista1())
```

```
len(lista1())
```

## Exemplos Big-O

- https://www.bigocheatsheet.com/

### Constant - O(1)

```
lista = [1, 2, 3, 4, 5]
```

```
def constant(n):
  print(n[0])
  print(n[1])
```

```
constant(lista)
```

### Linear - O(n)

```
def linear(n):
  for i in n:
    print(i)
```

```
linear(lista)
```

### Quadratic - O(n^2) - polynomial

```
def quadratic(n):
  for i in n:
    for j in n:
      print(i, j)
    print('---')  
```

```
quadratic(lista)
```

### Combination

```
# O(1) + O(5) + O(n) + O(n) + O(3)
# O(9) + O(2n) -> O(n)

def combination(n):
  # O(1) - Constante
  print(n[0])

  # O(5) - Constante
  for i in range(5):
    print('test ', i)

  # O(n) - Linear
  for i in n:
    print(i)

  # O(n) - Linear
  for i in n:
    print(i)

  # O(3) - Constante
  print('Python')
  print('Python')
  print('Python')
```

```
combination(lista)
```
