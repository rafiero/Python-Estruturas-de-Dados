```
def getKey(key):
  h = 0
  for char in key:
    h += ord(char)#ord returns an ASC value
  return h % 100
```

```
getKey("a")
```

```
class HashTable:
  def __init__(self):
    self.MAX = 100
    self.arr = [None for i in range(self.MAX)]

  def get_hash(self, key):
    h = 0
    for char in key:
      h += ord(char)#ord returns an ASC value
    return h % 100

  def add(self, key, value):
    h = self.get_hash(key)
    self.arr[h] = value
  
  def get(self, key):    
    return self.arr[self.get_hash(key)]

  #similar ao add
  def __setitem__(self, key, value):
    h = self.get_hash(key)
    self.arr[h] = value

  #similar ao get
  def __getitem__(self, key, value):
    h = self.get_hash(key)
    self.arr[h] = value
  
  def __delitem__(self, key):
    self.arr[self.get_hash(key)] = None
```

```
t = HashTable()
```

```
t.get_hash('Rafael')
```

```
t.add("Rafael", 10)
t.add("Rafaem", 100)
t.add("Rafaen", 10)
```

```
t.get("Rafael")
```
